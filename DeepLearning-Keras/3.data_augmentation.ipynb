### Data Augmentation occurs when we create new data based on existing data

Data Augmentation includes transformations like:
- Flipping the image horizontally or vertically
- Rotating the image
- Zooming in or out on the image
- Cropping the image
- Varying the color of the image

We do data augmentation to reduce overfitting

```python
import matplotlib.pyplot as plt
import numpy as np
import os
import random
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.preprocessing.image import ImageDataGenerator
%matplotlib inline
```

```python
def plotImages(images_arr):
    fig, axes = plt.subplots(1, 10, figsize=(20,20))
    axes = axes.flatten()
    for img, ax in zip( images_arr, axes):
        ax.imshow(img)
        ax.axis('off')
    plt.tight_layout()
    plt.show()
```

```python
gen = ImageDataGenerator(rotation_range=10, width_shift_range=0.1, height_shift_range=0.1, \
                         shear_range= 0.15, zoom_range=0.1, channel_shift_range=10., horizontal_flip=True)
```

```python
img_path = "../data/dogs-vs-cats/train/dog/dog.6.jpg"
```

```python
image = np.expand_dims(plt.imread(img_path), 0) # change the dimension to make the image compatible
```

```python
image.shape
```

```python
plt.imshow(image[0]);
```

```python
aug_iter = gen.flow(image)
```

```python
aug_images = [next(aug_iter)[0].astype(np.uint8) for i in range(10)]
```

```python
plotImages(aug_images);
```
